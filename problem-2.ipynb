Loading the npz file

```python
import numpy as np

# Load the npz file
data = np.load(r"C:\Users\DhananjayCiLE\gen_ai\3d_shape_points_data.npz")

# List all stored arrays
print(data.files)
```

Ploting the 3d shape

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

points = data["points"]

points.shape

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(points[:, 0], points[:, 1], points[:, 2], s=1, alpha=0.6)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("3D Shape Point Cloud")

plt.show()
```

Clustering using K-Means and reconstruction

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(points)

main_cluster = points[labels == np.bincount(labels).argmax()]

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(main_cluster[:, 0], main_cluster[:, 1], main_cluster[:, 2], s=1, alpha=0.6, color='orange')
ax.set_title("Reconstructed Shape (K-Means Clustering)")
plt.show()
```

The Idea behind this was to cluster the points using K means so that the defected points or the outlier points would be clustered into a cluster whereas the other points will be clustered into another one. The cluster with more points will obviously be the cluster which has thee correct points, hence the points from that cluster was used for the reconstruction, I admit that i have used chatgpt for some of the syntax.
